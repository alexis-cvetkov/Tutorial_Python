# 2. Matplotlib

Matplotlib is a Python 2D-plotting library that provides a Matlab-like syntax to visualize figures, histograms, images...

References: https://matplotlib.org/tutorials/index.html, http://www.labri.fr/perso/nrougier/teaching/matplotlib/ (2015, some features aren't supported anymore).

### Import Matplotlib

We will use the `pyplot` sub-module of the `matplotlib` package.

```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib notebook # Run this command to plot interactive figures
```

## 2.1Â General structure of a plotting script

- Initialization of the figure that will contain the plotted data with `plt.figure`, `plt.subplot`...

- Plotting of the data with `plt.plot`, `plt.scatter`, `plt.imshow`...

- Tuning of the figure (title, legends, axis labels...).

- Saving / Displaying of the figure

**Example:**

```python
# Create a new figure of size 8x6 points, using 80 dots per inch
plt.figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Plot cosine using blue color with a continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# Plot sine using green color with a continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Set x limits
plt.xlim(-4.0,4.0)

# Set x ticks
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# Set y limits
plt.ylim(-1.0,1.0)

# Set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# Save figure using 72 dots per inch
# savefig("./plt_example1.png",dpi=72)

# Show result on screen
plt.show()
```

**Quick plotting:**

If not defined, the figure and axes will be initialized with default parameters. This is useful for a quick visualization of the data.

```python
X = np.linspace(-np.pi, np.pi, 31,endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

##Â 2.2 1D plots with `plt.plot`

The `plt.plot` function is a must-know command of the matplotlib library. Basically, `plt.plot(x,y)` will plot `y` (an 1D array or list) as a function of `x` (1D array or list of same size as `y`).

Note: If `x` is not defined, `y` will be plotted against its indices.

Documentation: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

###Â Other parameters of the plot functions

In addition to `x`Â and `y` you can set many parameters in the `plt.plot` function to change the appearance of the curves (see):

- `color` (or c), string, set the color of the curve

```python
print('\n# --- Change colors')
plt.figure(figsize=(8,6), dpi=100)
plt.plot(X, C, color="red")
plt.plot(X, S, color="green")
plt.show()
```

- `linestyle` (or ls), `'-'` for a continuous line, `'--'` for a dashed line, `':'` for a dotted line and `''` for no line at all
- `linewidth` (or lw), float, the line width in points

```python
print('\n# --- Change linestyles / linewidths')
plt.figure(figsize=(8,6), dpi=100)
plt.plot(X, C, linestyle='--', linewidth=5)
plt.plot(X, S, linestyle='-', linewidth=2)
plt.show()
```

- `marker`, the marker of the data points, `'.'` for points, `'x'` for cross, `'+'` for plus. The markers are superimposed onto the linestyle of the curve.
- `markersize`, float, the size of the markers

```python
print('\n# --- Change markers / markersizes')
plt.figure(figsize=(8,6), dpi=100)
plt.plot(X, C, marker='o', markersize=7)
plt.plot(X, S, linestyle='', marker='+', markersize=7)
plt.show()
```

You can also use a condensed syntax to implement some of parameters detailed above.

```python
print('\n# --- Condensed syntax')
plt.figure(figsize=(8,6), dpi=100)
plt.plot(X, C, 'bo--') #Â 'b' -> Blue line, 'o' -> circle markers, '--' dashed line
plt.plot(X, S, 'r+') # 'b' -> Red line, '+' -> plus markers, no line
plt.show()
```

### Multiple plots

There are several ways to plot multiple sets of data on the same figure.

```python
print('\n# --- Use plt.plot several times')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
print('\n# --- plt.plot can take as input several arrays')
plt.plot(X, C, X, S)
plt.show()
```

```python
print('\n# --- Plot all the rows of an array\n')
x = np.arange(10)
A = np.arange(50).reshape((5,10))
print(f'A = \n{A}\n')
plt.plot(x,A.T) # plt.plot read A along rows so we need to transpose A
plt.show()
```

##Â 2.3 2D plots

### Plot images with imshow

`plt.imshow` allows you to plot arrays of shape (M,N) or (M,N,3) representing a grayscale or RGB image. Each "pixel" of the plotted image corresponds to a element of the array whose value set the color of the "pixel".

You can display the colorbar with `plt.colorbar` and change the colormap by setting the parameter `cmap` in `plt.imshow`.
The parameter `interpolation` sets the interpolation method for the visualization.

Doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html

```python
A = np.arange(25).reshape((5,5))
print(f'A = \n{A}\n')
plt.figure(figsize=(8,6), dpi=100)
plt.imshow(A, cmap='plasma', interpolation=None)
#plt.imshow(A, cmap='plasma', interpolation=None)
plt.colorbar()
plt.show()
```

## 2.4 Histograms plots

You can compute and plot 1D histograms with `n, bins, patches = plt.hist(A, bins=n_bins)`, where `A` is a 1D array and `n_bins` is the number of bins (divisions) of the histogram. Besides plotting the histogram of `A`, this function returns `n` (the number of data points in each bin) and `bins` (the values of the edges of ths bins).

Doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

```python
print('\n# --- 1D Histogram of a normal distribution')
plt.figure(figsize=(8,6), dpi=100)
A = np.random.randn(100000)
n,bins,_ = plt.hist(A,bins=100)
plt.show()
#print(bins)
```

Similarly you can use `n, x_edges, y_edges, image = plt.hist(A, B, bins=n_bins)` where `A` and `B` are 1D arrays  and `n_bins = [nx_bins,ny_bins]` is the number of bins along the x and y directions.

`n` is the number of data points in each bin, `x_edges` / `y_edges` are the bin edges along the x and y directions.

Use the command `plt.colorbar()` to display the colorbar.

Doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist2d.html

```python
print('\n# --- 2D Histogram of a normal distribution')
plt.figure(figsize=(8,6), dpi=100)
A = np.random.randn(100000)
B = np.random.randn(100000)
n,x_edges,y_edges,_ = plt.hist2d(A,B,bins=30)
plt.colorbar()
plt.show()
#print(bins)
```

## 2.5Â Figures, axes and subplots

###Â Figure creation

Use `plt.figure(fig_number)` to initialize a figure that will contain your plots. `fig_number` is an integer that is used as an ID for the figure. The curves plotted with `plt.plot` are automatically associated to the current figure.

- use the `figsize = (width, height)` parameter to set the size of the figure (in inches).
- use the `dpi` parameter to set the resolution (in dots per inches).

The total size of the figure is `(width*dpi, height*dpi)`.

```python
X = np.linspace(-np.pi, np.pi, 31,endpoint=True)
C,S = np.cos(X), np.sin(X)

#Â Figure 1
plt.figure(1,dpi=100)
plt.plot(X, C) #Â this curve is plotted in figure 1

# Figure 2
plt.figure(2,figsize=(10,8))
plt.plot(X, S) # this curve is plotted in figure 2

# Figure 1 again
#plt.figure(1)
#plt.plot(X, S) # this curve is plotted in figure 1

plt.show()
```

### Figure tuning 

- use the `label` parameter of `plt.plot` and the `plt.legend()` command to legend your plots.
- `plt.xlabel` / `plt.ylabel` to add labels to x,y axis.
- `plt.xscale('log')` / `plt.xscale('log')` to use log-scaled axis.
- `plt.title` to add a title to the figure.
- `plt.xlim` / `plt.ylim` to set the axis limits.
- `plt.xticks` / `plt.yticks` to set the axis ticks.

```python
x = np.linspace(0,1,100)
y = x.copy()
y_square = y ** 2

plt.figure(dpi=100)
plt.plot(x, y, label='y = x')
plt.plot(x, y_square, label='y = x^2')

#plt.xscale('log')
#plt.legend()
#plt.xlim(0,0.5)
#plt.xticks([0,0.25,0.5])
#plt.title('Title')
#plt.xlabel('x-label')
#plt.ylabel('y-label')

plt.show()
```

### Subplots grid

Use the `subplot`Â function to divide the main figure in smaller ones. `plt.subplot(n_rows, n_cols, k)` will add a subplot at theÂ `k`-th position in a grid of `n_rows` x `n_cols` subplots.

Call the function `plt.tight_layout` to reorganize the subplots so that they do not superimpose.

```python
fig = plt.figure(dpi=120)

plt.subplot(2,2,1)
plt.plot([1,2,3,4,5], [10,5,10,5,10], 'r-')
plt.title('Subplot 1')

plt.subplot(222) # Condensed format
plt.plot([1,2,3,4], [1,4,9,16], 'k-')
plt.title('Subplot 2')

plt.subplot(223)
plt.plot([1,2,3,4], [1,10,100,1000], 'b-')
plt.title('Subplot 3')

plt.subplot(224)
plt.plot([1,2,3,4], [0,0,1,1], 'g-')
plt.title('Subplot 4')

plt.tight_layout()
plt.show()
```

### Figure saving

Use `plt.savefig(filename,dpi)` to save the current figure under a certain `filename` and `dpi`.

```python
fig = plt.figure(dpi=120)

plt.subplot(2,2,1)
plt.plot([1,2,3,4,5], [10,5,10,5,10], 'r-')
plt.title('Subplot 1')

plt.subplot(222) # Condensed format
plt.plot([1,2,3,4], [1,4,9,16], 'k-')
plt.title('Subplot 2')

plt.subplot(223)
plt.plot([1,2,3,4], [1,10,100,1000], 'b-')
plt.title('Subplot 3')

plt.subplot(224)
plt.plot([1,2,3,4], [0,0,1,1], 'g-')
plt.title('Subplot 4')

plt.tight_layout()
plt.savefig('../data/example_subplots.png',dpi=120) #Â Save the current figure

plt.show()
```

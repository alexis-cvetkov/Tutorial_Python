### Curvefitting with scipy.optimize

We will use here the function `curve_fit` of the `scipy.optimize` package.

Parameters required for `curve_fit`:

- `f`, the function that will be optimized. `f` is assumed to be of the form: `f(xdata, param1, param2, ...)` where param1, param2, ... are the parameters to be fitted.
- `xdata`, array, the x-points on which the model will be optimized.
- `ydata`, array, the y-points on which the model will be optimized.
- `pO`, tuple, the guessed parameters of the model for the initialization.

Doc: https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html#scipy.optimize.curve_fit

```python
# Import the optimize subpackage of scipy
import scipy.optimize

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read experimental data
spectrum_1 = np.genfromtxt('../data/raman_spectrum.txt')
sigma, counts = spectrum_1[:,1], spectrum_1[:,2]

# Extract the part of the spectra for fitting
xfit = sigma[850:950]
yfit = counts[850:950]

# Plot the experimental data
plt.figure(dpi=120)
plt.plot(xfit,yfit)
plt.xlabel('sigma')
plt.ylabel('counts')
plt.title('Raman spectrum of Germanium')
plt.show()
```

```python
# Define the function whose parameters will be optimized. 

def gauss(x, A, mu, sigma, eps):
    return A*np.exp(-(x-mu)**2/(2.*sigma**2)) + eps
```

```python
# Optimize the model
popt, pcov = scipy.optimize.curve_fit(gauss,xfit,yfit,p0=(140,300,5,5))
A, mu, sigma, eps = popt
print('A = {}, mu = {}, sigma = {}, eps = {}'.format(A,mu,sigma,eps))
```

```python
# Compute fitted curve and plot it
y_opt = gauss(xfit, A, mu, sigma, eps)

plt.figure(dpi=120)
plt.plot(xfit,yfit,label='Experimental')
plt.plot(xfit,y_opt,label='Fitting')
plt.xlabel('sigma')
plt.ylabel('counts')
plt.title('Raman spectrum of Germanium')
plt.legend()
plt.show()
```
